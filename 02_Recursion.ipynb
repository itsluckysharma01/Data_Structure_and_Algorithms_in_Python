```python
## Example
 def fun(n):
    if n == 0:
        return 
    fun(n//2)
    print(n%2)
print(fun(13))
```

```python
# Print 1 to n
# n = 5
# out = 12345
def fun(n):
    if n == 0:
        return 
    fun(n-1)
    print(n, end=" ")

    

n = int(input("Enter Number: "))    
fun(n)
```

```python
# Print n to 1
# n = 5
# out = 54321
def fun(n):
    if n == 0:
        return
    print(n, end=" ")
    fun(n-1)
    

n = int(input("Enter Number: "))    
fun(n)
```

```python
#  Writing Base CAse in Recursion

# Factorial of Number

# 4 = 4*3*2*1 = 24

def fac(n):
    if n == 0:
        return 1
    return n*(fac(n-1))
    

n = int(input("Enter Number: "))
print(fac(n))
```
