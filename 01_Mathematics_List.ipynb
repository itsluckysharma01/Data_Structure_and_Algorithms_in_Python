```python
# Greatest Common Divisior

def  gcd(a,b):
    ag = set()
    bg = set()
    
    for i in range(1, a+1):
        if a%i == 0:
            ag.add(i)
    for i in range(1, b+1):
        if b%i == 0:
            bg.add(i)

    s = ag & bg
    if s != 1:
        return max(s)
    else:
        return 1
print(gcd(100,200))

            
```

```python
# LCM 

def lcm(a,b):
    res = max(a, b)
    while True:
        if res%a == 0 and res%b == 0:
            return res
        res +=1


print(lcm(8,6))
```

```python
# l = [10,20,30,40]
# output l =[20, 30, 40, 10]

def LeftRo(l):
    n = len(l)
    s = l[0]

    for i in range(1, n):
        l[i-1] = l[i]
    l[n-1] = s
    return l

    


l = [10,20,30,40]
print(LeftRo(l))
```
